```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dispersion(kx, E0, beta, a):
    return E0 - beta * (np.cos(kx * a) + 2)

def effmass(kx, beta, a):
    return 1 / (beta * np.cos(kx * a) * a**2)
```

```python
E0_V = 0
E0_L = 12
beta_V = -0.8
beta_L = 1.5
a = 4
kx = np.linspace(-np.pi / a, np.pi / a, 1000)

E_F = ((1.97**2)/2*511)*(3*np.pi**2/(a)**3)**(2/3)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(kx, dispersion(kx, E0_V, beta_V, a), label="Valenzelektronen")
plt.plot(kx, dispersion(kx, E0_L, beta_L, a), label="Leitungselektronen")
plt.hlines(E_F, -np.pi, np.pi, linewidth=1, color="red")
plt.xlabel('k_x * a')
plt.ylabel('E [eV]')
plt.title('Dispersionsrelationen von Valenzelektronen und Leitungselektronen')
#plt.yscale('log')
plt.legend()
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(kx, effmass(kx, beta_V, a), label="Effektive Masse der Valenzelektronen")
plt.xlabel('k_x * a')
plt.ylabel('E [eV]')
plt.title('Dispersionsrelationen von Valenzelektronen und Leitungselektronen')
#plt.yscale('log')
plt.legend()
```
